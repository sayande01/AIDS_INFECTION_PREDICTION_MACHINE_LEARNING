```python
# Assuming df is your DataFrame containing the column 'wtkg'
import scipy.stats as stats

plt.figure(figsize=(12, 6))

plt.subplot(121)
sns.distplot(df['time'])
plt.title('time Distribution')

plt.subplot(122)
stats.probplot(df['time'], dist="norm", plot=plt)
plt.title('time QQ Plot')

plt.show()
```

```python
from sklearn.preprocessing import QuantileTransformer
qt_normal = QuantileTransformer(output_distribution='normal')

df['time'] = qt_normal.fit_transform(df[['time']])
```

```python
# Assuming df is your DataFrame containing the column 'wtkg'
import scipy.stats as stats

plt.figure(figsize=(12, 6))

plt.subplot(121)
sns.distplot(df['time'])
plt.title('time Distribution')

plt.subplot(122)
stats.probplot(df['time'], dist="norm", plot=plt)
plt.title('time QQ Plot')

plt.show()
```

```python
# finding the outlier
percentile25=df["time"].quantile(0.25)
percentile75=df["time"].quantile(0.75)
# IQR CALCULATED
iqr=percentile75 - percentile25
iqr

# checking upper and lower limit in distance variable
upper_limit=percentile75 + 1.5*iqr # UPPER LIMIT
lower_limit=percentile25 - 1.5*iqr # LOWER LIMIT
print("upper_limit",upper_limit)
print("lower_limit",lower_limit)

# capping
new_df_cap = df.copy()

new_df_cap["time"]= np.where(
    new_df_cap["time"] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap["time"] < lower_limit,
        lower_limit,
        new_df_cap["time"]
    )

)
# Assign the modified "distance" column back to the original dataset

df['time'] = new_df_cap['time']
```
